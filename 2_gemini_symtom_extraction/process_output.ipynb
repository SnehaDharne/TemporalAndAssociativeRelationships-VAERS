```python
with open('output.txt', 'r') as f:
    data = f.readlines()
```

```python
data
```

```python
import json
data_points = []
for d in data:
    d= d.strip()
    json_object = json.loads(d)
    data_points.append(json_object)
```

```python
data_points
```

```python
data_for_df = []
for d in data_points:
    try:
        json_object = json.loads(d) 
        data_for_df.append(json_object)
    except json.JSONDecodeError as e:
        print(f"JSON decoding error: {e}")
        print(f"Error at position: {e.pos}")
    
    
```

```python
import pandas as pd
df = pd.DataFrame(data = data_for_df)
```

```python
data_for_df
```

```python
print("\nValue counts for symptom lists:")
print(df['extracted_symptoms'].value_counts().head(100))  # Top 10 most frequent symptoms
```

```python
df.to_csv('llm_tokenized_symptoms.csv')
```

```python
df['extracted_symptoms'] = df['extracted_symptoms'].apply(lambda x: [symptom.lower() for symptom in x])
df = df[df['extracted_symptoms'].apply(lambda x: len(x) > 0 and x != ['covid-19'])]
```
