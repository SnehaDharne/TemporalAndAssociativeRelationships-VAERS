```python
abbr = pd.read_csv('combined_medical_abbreviations_1.csv')
```

```python
abbr.head()
```

```python
abbr = abbr[['Abbreviation/Shorthand','Meaning']]
```

```python
abbr.head()
```

```python
abbreviations_dict = dict(zip(abbr['Abbreviation/Shorthand'], abbr['Meaning']))
```

```python
def replace_abbreviations(text, abbreviations_dict):
    words = text.split()  # Tokenize the text by splitting on whitespace
    new_words = [abbreviations_dict.get(word, word) for word in words]  # Replace if abbreviation exists
    return ' '.join(new_words)  # Join the text back together
```

```python
# Apply the function to replace abbreviations
symptoms_covid['SYMPTOM_TEXT'] = symptoms_covid['SYMPTOM_TEXT'].apply(lambda x: replace_abbreviations(x, abbreviations_dict))
```

```python
symptoms_covid['SYMPTOM_TEXT'] = symptoms_covid['SYMPTOM_TEXT'].str.replace(f'[{custom_punctuation}]', '', regex=True)
```

```python
import spacy
import pandas as pd

# Load the SciSpaCy model
nlp = spacy.load("en_ner_bc5cdr_md")
```

```python
def extract_symptoms(text):
    doc = nlp(text)
    symptoms = []
    for ent in doc.ents:
        # Check for entities labeled as "ENTITY"
        if ent.label_ == "ENTITY":
            symptoms.append(ent.text)
    return symptoms

# Apply the function to the SYMPTOM_TEXT column
symptoms_covid['Extracted_Symptoms'] = symptoms_covid['SYMPTOM_TEXT'].apply(extract_symptoms)

# Display some of the results
print(symptoms_covid[['SYMPTOM_TEXT', 'Extracted_Symptoms']].head())

```

```python
symptoms_covid['SYMPTOM_LIST']=''
```

```python
for i in range( len(symptoms_covid)):
    text = symptoms_covid['SYMPTOM_TEXT'].iloc[i]
    # Process the text
    doc = nlp(text)

    # Extract entities labeled as symptoms
    symptoms = [ent.text for ent in doc.ents if ent.label_ == "DISEASE"]
    print("Extracted Symptoms:", symptoms)
    symptoms_covid['SYMPTOM_LIST'][i] = symptoms
```
